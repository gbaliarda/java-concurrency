```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('matrix_multiplication_results_1024.csv')
```

```python
executor_df = df[df['Multiplication Model'].str.startswith('ExecutorService')]

executor_df['Threads'] = executor_df['Multiplication Model'].str.extract('(\d+)').astype(int)

executor_df = executor_df.sort_values('Threads')

sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))

plt.errorbar(
    executor_df['Threads'],
    executor_df['Average Execution Time (ms)'],
    yerr=executor_df['Standard Deviation (ms)'],
    fmt='-o',
    capsize=5,
    color='blue',
    label='ExecutorService'
)

plt.xlabel('Número de Hilos')
plt.ylabel('Tiempo Medio de Ejecución (ms)')
plt.title('Tiempo Medio de Ejecución: ExecutorService vs Número de Hilos')

plt.legend()

plt.tight_layout()
plt.show()
```

```python
forkjoin_df = df[df['Multiplication Model'].str.startswith('ForkJoin')]

forkjoin_df['Threads'] = forkjoin_df['Multiplication Model'].str.extract('(\d+)').astype(int)

forkjoin_df = forkjoin_df.sort_values('Threads')

sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))

plt.errorbar(
    forkjoin_df['Threads'],
    forkjoin_df['Average Execution Time (ms)'],
    yerr=forkjoin_df['Standard Deviation (ms)'],
    fmt='-o',
    capsize=5,
    color='blue',
    label='ExecutorService'
)

plt.xlabel('Número de Hilos')
plt.ylabel('Tiempo Medio de Ejecución (ms)')
plt.title('Tiempo Medio de Ejecución: ForkJoin vs Número de Hilos')

plt.legend()

plt.tight_layout()
plt.show()
```

```python
executor_df = df[df['Multiplication Model'].str.startswith('ExecutorService')]
executor_df['Threads'] = executor_df['Multiplication Model'].str.extract('(\d+)').astype(int)
executor_df = executor_df.sort_values('Threads')

forkjoin_df = df[df['Multiplication Model'].str.startswith('ForkJoin')]
forkjoin_df['Threads'] = forkjoin_df['Multiplication Model'].str.extract('(\d+)').astype(int)
forkjoin_df = forkjoin_df.sort_values('Threads')

sequential_row = df[df['Multiplication Model'] == 'Sequential']
sequential_value = sequential_row['Average Execution Time (ms)'].values[0]
sequential_error = sequential_row['Standard Deviation (ms)'].values[0]
sequential_min = sequential_value - sequential_error
sequential_max = sequential_value + sequential_error

sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))

plt.errorbar(
    executor_df['Threads'],
    executor_df['Average Execution Time (ms)'],
    yerr=executor_df['Standard Deviation (ms)'],
    fmt='-o',
    capsize=5,
    color='blue',
    label='ExecutorService'
)

plt.errorbar(
    forkjoin_df['Threads'],
    forkjoin_df['Average Execution Time (ms)'],
    yerr=forkjoin_df['Standard Deviation (ms)'],
    fmt='-o',
    capsize=5,
    color='green',
    label='ForkJoin'
)

plt.axhline(
    y=sequential_value,
    color='red',
    linestyle='--',
    label=f'Sequential ({sequential_value:.2f} ms)'
)

plt.fill_between(
    x=range(1, 21),
    y1=sequential_min,
    y2=sequential_max,
    color='red',
    alpha=0.2,
    label=f'Sequential Error ±{sequential_error:.2f} ms'
)

plt.xlabel('Número de Hilos')
plt.xlim(0.75,20.25)
plt.xticks(range(1,21))
plt.ylabel('Tiempo Medio de Ejecución (ms)')
plt.title('Comparación de Tiempos de Ejecución por Número de Hilos')
plt.legend()
plt.tight_layout()
plt.show()
```
